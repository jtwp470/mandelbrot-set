```python
import numpy as np
import pylab

image = np.zeros((320, 320), dtype=np.uint8)

width = image.shape[0]    # 幅   (x-axis)
height = image.shape[1]   # 高さ (y-axis)

for x in range(width):
    for y in range(height):
        image[x, y] = x + y

print(image)
pylab.imshow(image)
pylab.show()
```
